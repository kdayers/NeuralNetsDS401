```
from sklearn import datasets
from sklearn import linear_model

boston = sk.datasets.load_boston()
#model type #model <- lm( target ~. , data = boston)
model = sk.linear_model.LinearRegression()
fitted_model = model.fit(boston.data , boston.target)
prediction = fitted_model.predict( boston.data )
RMSE = np.sqrt( np.mean( (boston.target - prediction)**2 ) ) 
print("The RMSE is ", RMSE)

import matplotlib.pyplot as plt
```
